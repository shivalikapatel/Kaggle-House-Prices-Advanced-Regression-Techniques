## Feature Engineering

#### Visualizing Missing Data

```python
plt.figure(figsize = (15,7))
sns.heatmap(train.isnull(), cmap = 'viridis', cbar = False, yticklabels = False)
```

#### Removing all features with more than 50% missing values

```python
null_cols = train.isna().sum()>650
```

```python
null_cols[null_cols]
```

```python
train.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature', 'FireplaceQu'], axis =1 ,inplace = True)
```

```python
train.head()
```

```python
plt.figure(figsize = (15,7))
sns.heatmap(train.isnull(), cmap = 'viridis', cbar = False, yticklabels = False)
```

```python
cols = train.isna().sum()>0
```

```python
cls = cols[cols]
```

#### Replacing missing values of categorical columns with mode
#### Replacing missing values of numerical columns with mean

```python
for i in train.columns:
    if i in cls:
        if type(train[i][0]) == type('string'):
            train[i].fillna(train[i].mode()[0], inplace = True)
        else:
            train[i].fillna(train[i].mean(), inplace = True)
```

```python
sns.heatmap(train.isnull())
```

```python
train.drop(['Id'], axis = 1, inplace = True)
```

```python
train.head()
```

### Getting Modified Test Data

```python
test_df = pd.read_csv('test_data.csv')
```

```python
test_df.head()
```

### Combining Train and Test Data
#### Some categorical columns in Test Data have more values than Train Data
#### Combining Test and Train Data helps us to include those in our model

```python
comb_df = pd.concat([train, test_df], axis = 0)
```

```python
comb_df
```

```python
comb_df.reset_index(inplace = True)
```

```python
comb_df.drop(['index'],axis = 1, inplace = True)
```

```python
comb_df
```

### Extracting Categorical Columns from the combined data

```python
cols = []
for i in comb_df.columns:
    if type(comb_df[i][0]) == type('string'):
        cols.append(i)   
```

```python
cols
```

### Creating a function to get dummy variables for Categorical Columns

```python
def get_dummies(cat_cols):
    final_df = pd.DataFrame()
    i=0
    for col in cat_cols:
        dummy_df = pd.get_dummies(comb_df[col], drop_first = True)
        comb_df.drop([col],axis=1, inplace = True)
        if i == 0:
            final_df = dummy_df
        else:
            final_df = pd.concat([final_df, dummy_df], axis=1)

        i = i+1
    return pd.concat([comb_df, final_df], axis =1)
```

```python
df = get_dummies(cols)
```

```python
df
```

### Removing Duplicate Data

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df
```

### Dividing the Combined Data into Train and Test

```python
df_train = df.iloc[:1460, :]
```

```python
df_test = df.iloc[1460:, :]
```

```python
df_train.shape
```

```python
df_test.shape
```

### Removing the Dependent variable from Test Data

```python
df_test.drop(['SalePrice'], axis =1, inplace = True)
```

### Building a model to predict prices using XGBoost Regressor

```python
X_train = df_train.drop(['SalePrice'], axis =1)
y_train = df_train['SalePrice']
```

```python
import xgboost
model = xgboost.XGBRegressor()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(df_test)
```

```python
predictions
```

```python
test_pred = pd.DataFrame(predictions)
test_pred
```

### Saving our predictions in a sample_submission csv file

```python
sample_sub = pd.read_csv('sample_submission.csv')

sub = pd.concat([sample_sub['Id'], test_pred], axis =1)

sub.columns = ['Id', 'SalePrice']

sub.to_csv('sample_submission.csv', index = False)
```
